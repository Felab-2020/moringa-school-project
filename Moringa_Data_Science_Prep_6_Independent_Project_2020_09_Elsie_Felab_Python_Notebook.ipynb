Combining hours and minutes column

```
# Converting the columns hour and minute to string
hire = pd.read_csv('hire.csv', index_col= 0, dtype = {'hour':str, 'minute':str})
hire.head()
```

```
hire ['Time'] = hire.hour.str.cat([hire.minute], sep = ':')
hire
```

#1.3 Data Analysis

1. Identify the most popular hour of the day for picking up a shared electric car (Bluecar) in the city of Paris over the month of April 2018.

```
paris = hire[(hire.City=='Paris')]
paris.head()
```

```
# Popular hour of the day for picking shared electric blue car in Paris
paris.groupby(['hour'])['Bluecar counter'].sum()
```

2. What postal code is the most popular for picking up Blue cars?

```
paris.groupby(['Postal code','hour'])['Bluecar counter'].sum().sort_values(ascending = False).head()
```

3. What postal code is the most popular for picking up Blue cars?

```
paris.groupby(['Postal code','hour'])['Bluecar counter'].sum().sort_values(ascending = False).head()
```
